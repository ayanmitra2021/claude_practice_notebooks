```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
def add_user_message(messages, text):
    user_message = {"role":"user", "content":text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role":"assistant", "content":text}
    messages.append(assistant_message)


def chat(messages):  
    message = client.messages.create(
        model = model,
        max_tokens=1000,
        messages=messages
    )
    return message.content[0].text
```

```python
messages = [] 

add_user_message(messages, "Define quantum computing in one sentenece.")

answer = chat(messages)

add_assistant_message(messages, answer)

add_user_message(messages, "Write another sentenece.")

answer = chat(messages)

print(answer)
 
```

```python
all_messages = []
while True:
    user_message = input(">")
    print(">", user_message)
    add_user_message(all_messages, user_message)
    ai_response = chat(all_messages)
    add_assistant_message(all_messages, ai_response)       
    print("---- ", ai_response)
```
